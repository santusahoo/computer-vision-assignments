```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the image
img = cv.imread("/content/prac_picture_1.png")
```

```python
print(img.shape)
```

```python
gray_img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
print(gray_img.shape)
```

```python
resized_img = cv.resize(gray_img, (500,500), interpolation= cv.INTER_LINEAR)
```

```python
plt.imshow(resized_img)
plt.show()
```

```python
# Apply thresholding to the image to create a binary image
th, binary_img = cv.threshold(resized_img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
```

```python
print("Here the threshold value is",th)
```

```python
plt.imshow(binary_img)
plt.show()
```

##  Performing morphological operations to clean up the image

```python
# define the kernel size
kernel = np.ones((5, 5))
```

```python
# perform erosion
erosion = cv.erode(binary_img, kernel, iterations=1)
```

```python
# perform dilation
dilation = cv.dilate(binary_img, kernel, iterations=1)
```

```python
plt.title("binary image")
plt.imshow(binary_img)
plt.show()
```

```python
plt.title("Erosion")
plt.imshow(erosion)
plt.show()
```

```python
plt.title("Dilation")
plt.imshow(dilation)
plt.show()
```
